### Feature selection 

Select all samples(rows) that's been classified as C11

```python
X_rcv1['C11'] = y_rcv1['C11']
```

```python
X_rcv1_selected = X_rcv1[X_rcv1['C11']==1]
```

```python
X_rcv1_selected.shape
```

```python
X_rcv1_selected.head()
```

Count the number of appearance of each feature, then sort then in descending order

```python
selected_features = X_rcv1_selected.sum().sort_values(ascending=False)[:11]
```

Pick the best 10 features, cause these 10 features are most likely to contribute to C11

```python
selected_features
```

```python
X_rcv1 = X_rcv1[selected_features.index[1:]]
```
